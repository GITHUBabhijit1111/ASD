```python
import pandas as pd 
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

```python
Data=pd.read_csv('Saliva sampling.csv')
Data
```

```python
Data.rename(columns={'Imaginary Part':'ImaginaryMin','Unnamed: 3':'ImaginaryAvg','Real Part':'RealMin','Unnamed: 5':'RealAvg'},inplace=True)
```

```python
Data
```

```python
Data=Data.dropna()
```

```python
Data
```

```python
Data.drop(Data.columns[1],axis=1,inplace=True)
```

```python
Data
```

```python
Data.dtypes
```

```python
Data.ImaginaryMin=pd.to_numeric(Data.ImaginaryMin)
```

```python
Data.ImaginaryAvg=pd.to_numeric(Data.ImaginaryAvg)
```

```python
Data.RealMin=pd.to_numeric(Data.RealMin)
```

```python
Data.RealAvg=pd.to_numeric(Data.RealAvg)
```

```python
Data.dtypes
```

```python
Data=pd.get_dummies(data=Data,columns=["Smoking"])
Data
```

```python
Data.rename(columns={"Smoking_1.0":"Non_Smoker","Smoking_2.0":"Ex_Smoker","Smoking_3.0":"Active_Smoker"},inplace=True)
Data
```

```python
Data['Diagnosis'].replace({'COPD':0,'HC':1,'Asthma':2,'Infected':3},inplace=True)
Data
```

```python
Data=Data.drop(Data[Data['Diagnosis']==2].index)
Data=Data.drop(Data[Data['Diagnosis']==3].index)
```

```python
Data
```

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np
scaler=MinMaxScaler()
```

```python
reshaped_ImaginaryMin=pd.DataFrame(Data.ImaginaryMin)
Trans_ImaginaryMin=scaler.fit_transform(reshaped_ImaginaryMin)
```

```python
reshaped_ImaginaryAvg=pd.DataFrame(Data.ImaginaryAvg)
Trans_ImaginaryAvg=scaler.fit_transform(reshaped_ImaginaryAvg)
```

```python
reshaped_RealMin=pd.DataFrame(Data.RealMin)
Trans_RealMin=scaler.fit_transform(reshaped_RealMin)
```

```python
reshaped_RealAvg=pd.DataFrame(Data.RealAvg)
Trans_RealAvg=scaler.fit_transform(reshaped_RealAvg)
```

```python
Data["ImaginaryMin"]=Trans_ImaginaryMin.reshape(-1)
Data["ImaginaryAvg"]=Trans_ImaginaryAvg.reshape(-1)
Data["RealMin"]=Trans_RealMin.reshape(-1)
Data["RealAvg"]=Trans_RealAvg.reshape(-1)
```
